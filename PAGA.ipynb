# Run the algorithm

```python
sc.tl.pca(ad, svd_solver='arpack')
sc.pp.neighbors(ad)
sc.tl.draw_graph(ad)

sc.pl.draw_graph(ad, legend_loc='on data', color = 'merged')
sc.tl.leiden(ad)

sc.tl.paga(ad, groups='merged')

sc.pl.paga(ad)

#ad.uns['iroot'] = np.flatnonzero(ad.obs['merged']  == 3)[0]
ad.uns['iroot'] = np.flatnonzero(ad.obs['merged'] == '7')[0]
#ad.uns['iroot'] = 0

sc.tl.dpt(ad)

sc.pl.draw_graph(ad, color=[str('merged'), 'dpt_pseudotime'], legend_loc='on data')

#scvelo.pl.paga(ad)
```
