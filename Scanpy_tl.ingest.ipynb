# Load in 

```python
import scanpy as sc
from mousipy import translate
import pandas as pd
```

# Import data

```python
adata_ref = sc.read_h5ad('/Volumes/hqdinh2/Projects/HNC_SPORE/Golfinosetal2022/Seurat_label_transfer_mouse_to_human/tl.Ingest/human_reference_downsampled.h5ad')
adata = sc.read_h5ad('/Volumes/hqdinh2/Projects/HNC_SPORE/Golfinosetal2022/Seurat_label_transfer_mouse_to_human/tl.Ingest/mouse_query.h5ad')
```

# Convert mouse object to human

```python
humanized_adata = translate(adata)
```

```python
humanized_adata
```

# Define both datasets on the same variables

```python
var_names = adata_ref.var_names.intersection(humanized_adata.var_names)
adata_ref = adata_ref[:, var_names]
humanized_adata = humanized_adata[:, var_names]
adata_ref
```

# Dimension reduction

```python
sc.pp.pca(adata_ref)
sc.pp.neighbors(adata_ref)
sc.tl.umap(adata_ref)
```

# Plotting UMAP

```python
sc.pl.umap(adata_ref, color='global.cluster4')
```

# Run tl.ingest

```python
sc.tl.ingest(humanized_adata, adata_ref, obs='global.cluster4')
#humanized_adata.uns['colors'] = adata_ref.uns['global.cluster4']  # fix colors
```
